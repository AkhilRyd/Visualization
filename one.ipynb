```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y,marker='o', linestyle='--', color='b', label='Linear Growth')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Categories and their respective values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 25, 20, 25]
```

```python
import matplotlib.pyplot as plt

# Pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%',startangle=90)
plt.title('Basic Pie Chart')
plt.show()
```

```python
colors = ['gold', 'skyblue', 'lightcoral', 'yellowgreen']
explode = [0.1, 0, 0, 0]  # "Explode" the first slice

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90,
        colors=colors, explode=explode, shadow=True)
plt.title('Customized Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for a single violin plot
data = np.random.normal(0, 1, 100)  # 100 random numbers from a normal distribution

# Create the violin plot
plt.violinplot(data, showmeans=True, showmedians=True)

# Customizing the plot
plt.title('Single Violin Plot')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.xticks([1], ['Dataset'])  # Label x-axis

# Show the plot
plt.show()
```
